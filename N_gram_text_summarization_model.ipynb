```python
from PyPDF2 import PdfReader
```

```python
def parse_pdf_to_text(pdf_path):
    pdf_reader = PdfReader(pdf_path)
    text = ''
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        text += page.extract_text()
    return text

# original_size = len(text)
# print("Size of given text : ")
# print(original_size)
text = parse_pdf_to_text("Construction_Articles.pdf")
text
```

```python
# creating tokens 
token = text.replace("\n", " ").split()
```

```python
def create_ngram_model(text, n):
    # create empty dictionary
    n_gram_model = {}
    
    # loop through given text data with window of size n upto len(text) - n
    for i in range(len(text)-n):
        # creating tuple of n_grams of size 'n' and assign as a key to n_gram_model
        n_gram_key = tuple(text[i:i+n])
        
        # next word
        next_word = text[i+n]
        
        # checking for key in dictionary if not present then create new empty dictionary for that key
        if n_gram_model.get(n_gram_key) == None:
            n_gram_model[n_gram_key] = {}
        
        n_gram_model[n_gram_key][next_word] = n_gram_model[n_gram_key].get(next_word, 0) + 1
    
    
    # calculating relative probabilities for each possible words
    
    for ngram, next_words in n_gram_model.items():
        total = sum(next_words.values())
        for word in next_words :
            n_gram_model[ngram][word] /= total
            
    return n_gram_model 
```

```python
import random

def generate_ngram_text(ngram_model, start_words, length):
    generated_text = list(start_words)
    for _ in range(length):
        current_ngram = tuple(generated_text[-N:])
#         print(current_ngram)
        if current_ngram in ngram_model:
            next_word_probs = ngram_model[current_ngram]
            next_word = random.choices(list(next_word_probs.keys()), list(next_word_probs.values()))[0]
#             print(random.choices(list(next_word_probs.keys())))
            generated_text.append(next_word)
        else:
            break
    result = " ".join(generated_text)
    return result
```
